```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
col_names = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety', 'class_value']
df = pd.read_csv("car.data", names=col_names)
```

```python
label_encoders={}
for column in ['class_value']:
    encoder = LabelEncoder()
    df[column] = encoder.fit_transform(df[column])
    label_encoders[column]=encoder
```

```python

# Kategorik değişkenler ve hedef değişken arasındaki ilişkiyi görselleştirme
categorical_variables = ['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety']
target_variable = 'class_value'
```

```python
for column in categorical_variables:
    plt.figure(figsize=(8, 6))
    sns.countplot(x=column, hue=target_variable, data=df)
    plt.title(f'{column} vs {target_variable}')
    plt.xlabel(column)
    plt.ylabel('Count')
    plt.legend(title=target_variable)
    plt.show()
```
